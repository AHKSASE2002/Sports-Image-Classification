```python
import pandas as pd
import torch
import os
import matplotlib.pyplot as plt
```

```python
paths = [
    "./logs_128_pretrained/checkpoints",
    "./logs_224_pretrained/checkpoints",
    "./logs_224_scratch/checkpoints",
]
```

```python
metrics_dict = {

}
for path in paths:
    models_paths = os.listdir(path)
    for model_path in models_paths:
        full_path = os.path.join(path, model_path)
        print(full_path)
        metrics_dict[full_path] = torch.load(full_path+"/metrics.pt", map_location=torch.device('cpu'))
```

```python
for full_path, metrics in metrics_dict.items():
    epochs = range(1, len(metrics['train_losses']) + 1)
    model_name = full_path.split("\\")[-1]
    # === Plot 1: Loss ===
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, metrics['train_losses'], label='Train Loss')
    plt.plot(epochs, metrics['val_losses'], label='Validation Loss')
    plt.title(f'Loss Over Epochs - {model_name}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'{full_path}/loss_plot.png')

    # === Plot 2: Accuracy ===
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, metrics['train_accs'], label='Train Accuracy')
    plt.plot(epochs, metrics['val_accs'], label='Validation Accuracy')
    max_val_acc = max(metrics['val_accs'])
    max_val_acc_epoch = metrics['val_accs'].index(max_val_acc) + 1
    plt.title(f'Accuracy Over Epochs - {model_name}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'{full_path}/accuracy_plot.png')
    plt.close('all')
```
